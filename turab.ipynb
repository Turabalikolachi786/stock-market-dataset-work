```python
import numpy as np
import pandas as od
```

```python
data= pd.read_csv("Amazon.csv")
```

```python
data = np.array([
    ["2023-01-01", 100, 120, 90, 110, 115, 1000],
    ["2023-01-02", 105, 125, 95, 115, 120, 2000],
    ["2023-01-03", 110, 130, 100, 120, 125, 1500],
    ["2023-01-04", 115, 135, 105, 125, 130, 3000],
    ["2023-01-05", 120, 140, 110, 130, 135, 4000],
], dtype=object)
```

```python
data
```

```python
dates = data[:, 0]  # Date column
numeric_data = data[:, 1:].astype(float)
numeric_data
```

```python
open_prices = numeric_data[:, 0]  # Open column
close_prices = numeric_data[:, 3]  # Close column
print("Open Prices:", open_prices)
print("Close Prices:", close_prices)
```

```python
first_two_rows = numeric_data[:2, :]  # First 2 rows
specific_columns = numeric_data[:, [0, 3]]  # Open and Close columns
print("First 2 Rows:\n", first_two_rows)
print("Open and Close Columns:\n", specific_columns)
```

```python
mean_close = close_prices.mean()  # Mean of Close prices
max_volume = numeric_data[:, -1].max()  # Maximum Volume
min_low = numeric_data[:, 2].min()  # Minimum Low price
print("Mean Close:", mean_close)
print("Max Volume:", max_volume)
print("Min Low:", min_low)
```

```python
filtered_rows = numeric_data[numeric_data[:, 0] > 110]  # Rows where Open > 110
print("Filtered Rows:\n", filtered_rows)
```

```python
numeric_data[numeric_data[:, -1] == 3000, -1] = np.nan
print("Replaced Volume 3000 with NaN:\n", numeric_data)
```

```python
numeric_data[np.isnan(numeric_data)] = 0  # Replace NaN with 0
print("Replaced NaN with 0:\n", numeric_data)
```

```python
diff_high_low = numeric_data[:, 1] - numeric_data[:, 2]
numeric_data = np.column_stack((numeric_data, diff_high_low))
print("Added Difference Column:\n", numeric_data)
```

```python
sorted_by_close = numeric_data[numeric_data[:, 3].argsort()]
print("Sorted by Close Prices:\n", sorted_by_close)
```

```python
total_volume = numeric_data[:, -2].sum()  # Summing Volume column
print("Total Volume:", total_volume)
```

```python
unique_open_prices = np.unique(open_prices)
print("Unique Open Prices:", unique_open_prices)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
da= pd.read_csv("Amazon.csv")
```

```python
da = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"],
    "Open": [100, 105, 110, 115, 120],
    "High": [120, 125, 130, 135, 140],
    "Low": [90, 95, 100, 105, 110],
    "Close": [110, 115, 120, 125, 130],
    "Adj Close": [115, 120, 125, 130, 135],
    "Volume": [1000, 2000, 1500, 3000, 4000],
}
```

```python
da
```

```python
df = pd.DataFrame(da)
df
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df["High"], bins=5, color='r', edgecolor='black')
plt.xlabel("High Prices")
plt.ylabel("Frequency")
plt.title("Distribution of High Prices")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["High"], df["Low"], color='purple')
plt.xlabel("High Prices")
plt.ylabel("Low Prices")
plt.title("Scatter Plot of High vs Low Prices")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Open"], label="Open Price", color='c', marker='x')
plt.plot(df["Date"], df["Close"], label="Close Price", color='m', marker='o')
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Open vs Close Prices Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Open"], label="Open Price", color='c', marker='x')
plt.plot(df["Date"], df["Close"], label="Close Price", color='m', marker='o')
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Open vs Close Prices Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Close"], label="Close Price", color='b', marker='o')
plt.title("Customized Line Plot of Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
